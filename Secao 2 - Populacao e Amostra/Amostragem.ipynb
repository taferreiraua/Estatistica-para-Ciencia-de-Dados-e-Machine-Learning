## **Amostragem Sistemática**

* Seleciona-se o primeiro elemento para a amostra e os itens seguintes são escolhidos utilizando intervalos fixos ou sistemáticos até chegar ao tamanho da amostra desejada.

O valor para o intervalo é dado pela divisão do tamanho da população pelo tamanho da amostra desejada:

```python
intervalo = len(dataset)//100
```

```python
random.seed(1) # retorna sempre o mesmo valor
inicio = random.randint(0,intervalo)
```

```python
np.arange(inicio, len(dataset), step=intervalo)
```

```python
def amostragem_sistematica(dataset, amostras):
  random.seed(1)

  intervalo = len(dataset)//amostras
  inicio = random.randint(0, intervalo)
  indices = np.arange(inicio, len(dataset), step=intervalo)

  amostra_sistematica = dataset.iloc[indices]

  return amostra_sistematica
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100)
df_amostra_sistematica.shape
df_amostra_sistematica.head()
```

## **Amostra por conglomerado (grupos)**

* A população inteira é dividida em grupos(clusters) distintos, de modo que não difiram quanto a caracteristica desejada. Em seguida, seleciona-se aleatóriamente alguns deles para estudo.

Primeiro define-se o número de grupos (tamanho da população dividido pelo numero desejado de grupos):

```python
intervalo = len(dataset)//10
```

```python
grupos = list()
id_gp = 0
count = 0

for _ in dataset.iterrows() :
  grupos.append(id_gp)
  count += 1
  if count > intervalo:
    count = 0
    id_gp += 1
```

Elementos por grupo:

```python
np.unique(grupos, return_counts=True)
```

O tamanho da lista de grupos deve ser o mesmo do dataset.

```python
np.shape(grupos), dataset.shape
```

Concatenando 'grupos' e 'dataset' e escolhendo o grupo:

```python
dataset['grupo'] = grupos
dataset.head()
```

```python
random.seed(1)
random_gp = random.randint(0,9)
```

```python
df_agrupamento = dataset[dataset['grupo'] == random_gp]
df_agrupamento['grupo'].value_counts()
```

```python
def amostragem_conglomerado(dataset, ngrupos):
  random.seed(1)

  intervalo = len(dataset)//ngrupos

  grupos = list()
  id_gp = 0
  count = 0
  for _ in dataset.iterrows() :
    grupos.append(id_gp)
    count += 1
    if count > intervalo:
      count = 0
      id_gp += 1

  dataset['grupo'] = grupos
  random_gp = random.randint(0, ngrupos)
  return dataset[dataset['grupo'] == random_gp]
```

```python
df_amostra_conglomerado = amostragem_conglomerado(dataset, 100)
df_amostra_conglomerado.shape, df_amostra_conglomerado['grupo'].value_counts()
```

## **Amostragem Estratificada**

* Consiste em especificar quantos elementos da amostra serão retirados de cada estrato, de modo proporcional.

'Test_size' para uma amostra de 100 registros:

```python
test_size = 100/len(dataset)
```

O codigo abaixo vai gerar 2 datasets: um com 10% da população e um com os outros 90%:

```python
split = StratifiedShuffleSplit(test_size=test_size) # indica o tamanho da amostra (nesse caso, 10%)
for x, y in split.split(dataset, dataset['income']):
  df_90 = dataset.iloc[x]
  df_10 = dataset.iloc[y]

amostra = df_10

df_90.shape, df_10.shape
```

```python
amostra['income'].value_counts()
```



> **Obs.:** Em Machine Learning, podemos chamar o dataset correspondente aos 90% de "base de treino" e o de 10% de "base de teste".



```python
amostra.head()
```

```python
def amostragem_estratificada(dataset, variavel, percent):
  split = StratifiedShuffleSplit(test_size=percent, random_state=1)
  for treino, teste in split.split(dataset, dataset[variavel]):
    treino = dataset.iloc[treino]
    teste = dataset.iloc[teste]

  return treino, teste
```

```python
treino, teste = amostragem_estratificada(dataset, 'income', test_size)
teste.shape
```

```python
df_amostra_estratificada = teste
```

## **Amostra de Reservatório**
* Sorteiam-se itens de mesma probabilidade de seleção do stream de dados.
* Utilizada quando trabalhamos com Data stream(dados não estáticos) de itens com tamanho desconhecido que podem ser acessados somente uma vez.

**Ex.: Retiramos um mostra de garrafas plásticas da esteira de uma fábrica.**

```python
stream = list()
for i in range(len(dataset)) :
  stream.append(i)
```

```python
def amostragem_reservatorio(dataset, nAmostras):
  stream = list()
  
  for i in range(len(dataset)):
    stream.append(i)

  i = 0
  tamanho = len(dataset)

  reservatorio = [0] * nAmostras

  for i in range(nAmostras):
    reservatorio[i] = stream[i]

  while i < tamanho:
    j = random.randrange(i + 1)
    if j < nAmostras:
      reservatorio[j] = stream[i]
    i += 1

  return dataset.iloc[reservatorio]
```

```python
df_amostragem_reservatorio = amostragem_reservatorio(dataset, 100)
df_amostragem_reservatorio.shape
```

```python
df_amostragem_reservatorio.head()
```

Comparando a representatividade dos resultados quanto a media das variáveis 'age', 'income' e 'loan':

```python
def comparar_resultados(comparativo, variavel):
  amostras = {
    'Aleatoria': abs(comparativo - amostra_aleatoria[variavel].mean()),
    'Sistematica': abs(comparativo - amostra_sistematica[variavel].mean()),
    'Conglomerada': abs(comparativo - amostra_conglomerado[variavel].mean()),
    'Reservatorio': abs(comparativo - amostra_reservatorio[variavel].mean()),
    'Estratificada': abs(comparativo - amostra_estratificada[variavel].mean())
  }
  return min(amostras, key=amostras.get)


age = comparar_resultados(df['age'].mean(), 'age')
print(f'Quanto ao parâmetro "age": Amostra {age}')

income = comparar_resultados(df['income'].mean(), 'income')
print(f'Quanto ao parâmetro "income": Amostra {income}')

loan = comparar_resultados(df['loan'].mean(), 'loan')
print(f'Quanto ao parâmetro "loan": Amostra {loan}')
```
